## Scale

```python
from sklearn.preprocessing import scale
```

```python
crime_data_scaled = scale(crime_data)
crime_data_scaled  = pd.DataFrame(crime_data_scaled,columns=crime_data.columns)
```

```python
crime_data_scaled
```

## Hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,linkage
from scipy.spatial.distance import pdist
```

### Dendogram

```python
xdist  = pdist(crime_data_scaled,metric='euclidean')
linked = linkage(xdist,method='ward')
plt.figure(figsize=[15,7])
dendrogram(linked,orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
hierachichal = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
pred_hierachical = hierachichal.fit_predict(crime_data_scaled)
pred_hierachical
crime_data['Hierachical'] = pred_hierachical
crime_data
```

```python
plt.scatter(x=crime_data['Hierachical'],y=crime_data['UrbanPop'],c=pred_hierachical)
plt.show()
```

## Kmeans

```python
from sklearn.cluster import KMeans
```

### Optimum kmeans cluster

```python
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(crime_data_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

### Obersvation

## 4 is best optimum cluster

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(crime_data_scaled)
pred_kmeans = kmeans.predict(crime_data_scaled)
crime_data['kmeans'] = pred_kmeans
pred_kmeans
crime_data
```

```python
plt.scatter(x=crime_data['kmeans'],y=crime_data['UrbanPop'],c=pred_kmeans)
plt.show()
```
